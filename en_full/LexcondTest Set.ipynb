```python
import csv
from nltk.tokenize.moses import MosesTokenizer
import pprint
from collections import Counter, defaultdict
import _pickle as cPickle
from tqdm import tqdm

tknzr = MosesTokenizer()
pp = pprint.PrettyPrinter(indent=4)
reader = csv.DictReader(open('trainset.csv', 'rt', encoding='utf-8'))
#writer = csv.DictWriter(open('devset_located.csv', 'wt', encoding='utf-8', newline='\n'), fieldnames=['mr', 'ref'])
#writer.writeheader()
headers = ['name', 'eatType','priceRange', 'customer rating', 'near','food','area','familyFriendly']
ovocabulary = defaultdict(lambda: [])
feature_vocabulary = defaultdict(lambda: defaultdict(lambda: Counter()))
name = ''
food = 'adsddass'
feature_to_startword = defaultdict(lambda: Counter())
for row in tqdm(reader):
    i1 = row['mr']
    i2 = row['ref']
    
    keywords = i1.split(',')
    kv_map = {}
    for keyword in keywords:
        k_idx = keyword.find('[')
        k = keyword[:k_idx].strip()
        value = keyword[k_idx + 1: keyword.find(']')]
        ovocabulary[k].append(value)
        kv_map[k] = value
        
        if k == 'name':
            i2 = i2.replace(value, 'XNAMEX')
        if k == 'food':
            i2 = i2.replace(value, 'XFOODX')
    feature = ''
    for header in headers:
        if header in kv_map.keys():
            feature += '1'
        else:
            feature += '0'    
    
    
    tokens = tknzr.tokenize(i2)
    first_word = tokens[0]
    feature_to_startword[feature].update([first_word])
    for keyword, value in kv_map.items(): 
        feature_vocabulary[keyword][value].update([tokens[0]])
    
    #writer.writerow({'mr': i1, 'ref': 'Located in'})   

```

```python
pp = pprint.PrettyPrinter(indent=4)
start_word_to_feature = defaultdict(lambda: defaultdict(lambda: 0))
for ft, words in feature_to_startword.items():
    for word, freq in words.items():
        if freq > 5:
            start_word_to_feature[word][ft] += freq
pp.pprint(start_word_to_feature)
```
