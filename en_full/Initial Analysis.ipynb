```python
import csv


reader = csv.DictReader(open('trainset.csv', 'rt', encoding='utf-8'))

lengths1 = []
lengths2 = []
for row in reader:
    i1 = row['mr']
    i2 = row['ref']
    
    lengths1.append(len(i1))
    lengths2.append(len(i2))
    
print('Max Input Length: {}'.format(max(lengths1)))
print('Max Output Length: {}'.format(max(lengths2)))

print('Min Input Length: {}'.format(min(lengths1)))
print('Min Output Length: {}'.format(min(lengths2)))

print('Mean Input Length: {}'.format(sum(lengths1)/len(lengths1)))
print('Mean Output Length: {}'.format(sum(lengths2)/len(lengths2)))
```

```python
from nltk.tokenize import WordPunctTokenizer
from collections import defaultdict


reader = csv.DictReader(open('trainset.csv', 'rt', encoding='utf-8'))
tokenizer = WordPunctTokenizer()

ovocabulary =  defaultdict(lambda: 0)
for row in reader:
    #i1 = tokenizer.tokenize(row['mr'])
    i2 = tokenizer.tokenize(row['ref'])

    for tok in i2:
        ovocabulary[tok] += 1

print(len(ovocabulary))

sorted_vocab = sorted(ovocabulary.items(), key=lambda x: x[1], reverse=True)
print(sorted_vocab)
```

```python
from collections import defaultdict
import csv

reader = csv.DictReader(open('trainset.csv', 'rt', encoding='utf-8'))

ovocabulary =  defaultdict(lambda: [])
for row in reader:
    i1 = row['mr']
    keywords = i1.split(',')
    for keyword in keywords:
        k_idx = keyword.find('[')
        k = keyword[:k_idx].strip()
        value = keyword[k_idx + 1: keyword.find(']')]
        ovocabulary[k].append(value)

sorted_vocab = sorted(ovocabulary.items(), key=lambda x: len(x[1]), reverse=True)
print(ovocabulary.keys())
print({k: len(v) for k,v in ovocabulary.items()})
print({k: len(set(v)) for k,v in ovocabulary.items()})
```
